# Importando biblioteca

```
import pandas as pd
import numpy as np
import random
```

#Setup

```
# Processo A
Entrada_A = random.randint(100, 1000)
Refugo_A = int(Entrada_A * random.uniform(0.01, 0.05))
Saida_A = Entrada_A - Refugo_A
Defeitos_A = int(Entrada_A * random.uniform(0.005, 0.02))
Retrabalho_A = int(Entrada_A * random.uniform(0.005, 0.02))

# Processo B
Entrada_B = Saida_A
Refugo_B = int(Entrada_B * random.uniform(0.01, 0.05))
Saida_B = Entrada_B - Refugo_B
Defeitos_B = int(Entrada_B * random.uniform(0.005, 0.02))
Retrabalho_B = int(Entrada_B * random.uniform(0.005, 0.02))

# Processo C
Entrada_C = Saida_B
Refugo_C = int(Entrada_C * random.uniform(0.01, 0.05))
Saida_C = Entrada_C - Refugo_C
Defeitos_C = int(Entrada_C * random.uniform(0.005, 0.02))
Retrabalho_C = int(Entrada_C * random.uniform(0.005, 0.02))
```

```
# Organização em lista de processos
processos = [
    {"processo": "A", "entrada": Entrada_A, "saida": Saida_A, "refugo": Refugo_A, "defeitos": Defeitos_A, "retrabalho": Retrabalho_A},
    {"processo": "B", "entrada": Entrada_B, "saida": Saida_B, "refugo": Refugo_B, "defeitos": Defeitos_B, "retrabalho": Retrabalho_B},
    {"processo": "C", "entrada": Entrada_C, "saida": Saida_C, "refugo": Refugo_C, "defeitos": Defeitos_C, "retrabalho": Retrabalho_C}
]
```

## Visualização da saída

```
for processo in processos:
    print(f"Processo: {processo['processo']}")
    print(f"Entrada: {processo['entrada']}")
    print(f"Saida: {processo['saida']}")
    print(f"Refugo: {processo['refugo']}")
    print(f"Defeitos: {processo['defeitos']}")
    print(f"Retrabalho: {processo['retrabalho']}")
    print()
```

## FTY

```
print("Calculando os FTY")

#porcentagem de itens que passam no processo sem a necessidade de retrabalho ou refugo na primeira vez que são processados

FTY_A = (Entrada_A-Refugo_A)/Entrada_A*100
print(f'FTY A é de: {FTY_A:.2f}% \n')

print("Calculando o FTY")
FTY_B = (Entrada_B-Refugo_B)/Entrada_B*100
print(f'FTY B é de: {FTY_B:.2f}% \n')

print("Calculando o FTY")
FTY_C = (Entrada_C-Refugo_C)/Entrada_C*100
print(f'FTY C é de: {FTY_C:.2f}% \n')
```

## FPY

```
print("Calculando os FPY")

#porcentagem de itens que são aprovados ao final do processo, considerando refugos e retrabalhos

FPY_A = (Entrada_A-Refugo_A-Retrabalho_A)/Entrada_A*100
print(f'FPY A é de: {FPY_A:.2f}% \n')

FPY_B = (Entrada_B-Refugo_B-Retrabalho_B)/Entrada_B*100
print(f'FPY B é de: {FPY_B:.2f}% \n')

FPY_C = (Entrada_C-Refugo_C-Retrabalho_C)/Entrada_C*100
print(f'FPY C é de: {FPY_C:.2f}% \n')
```

```
## RTY
```

```
print("Calculando os RTY")

#porcentagem de itens que saem do processo sem defeitos em relação à quantidade total de entrada

RTY_A = (Entrada_A-Defeitos_A)/Entrada_A*100
print(f'RTY A é de: {RTY_A:.2f}% \n')

RTY_B = (Entrada_B-Defeitos_B)/Entrada_B*100
print(f'RTY B é de: {RTY_B:.2f}% \n')

RTY_C = (Entrada_C-Defeitos_C)/Entrada_C*100
print(f'RTY C é de: {RTY_C:.2f}% \n')
```

## RTY TOTAL

```
print("Calculando o RTY TOTAL")

# medição da qualidade geral do processo
RTY_TOTAL = (FTY_A / 100) * (FTY_B / 100) * (FTY_C / 100)

print(f'RTY TOTAL é de: {RTY_TOTAL * 100:.2f}% \n')
```
